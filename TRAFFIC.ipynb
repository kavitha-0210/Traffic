```python
# TRAFFIC PATTERN ANALYSIS (Vehicle Count per Hour)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# -------------------------------------------------------
# 1. Load Dataset
# -------------------------------------------------------

# Example CSV format:
# Hour,Vehicle_Count
# 0,120
# 1,90
# 2,70
# ...
df = pd.read_csv("traffic_data.csv")

print("Dataset Loaded:")
print(df.head())

# -------------------------------------------------------
# 2. Check for missing values
# -------------------------------------------------------
df = df.dropna()

# -------------------------------------------------------
# 3. Plot Traffic Density Over Time
# -------------------------------------------------------
plt.figure(figsize=(10,5))
plt.plot(df['Hour'], df['Vehicle_Count'], marker='o')
plt.title("Traffic Density over Time")
plt.xlabel("Hour of Day")
plt.ylabel("Vehicle Count")
plt.grid(True)
plt.show()

# -------------------------------------------------------
# 4. Identify Peak Traffic Hours
# -------------------------------------------------------
peak = df.loc[df['Vehicle_Count'].idxmax()]

print("\nPeak Traffic Hour:", peak['Hour'])
print("Vehicle Count:", peak['Vehicle_Count'])

# -------------------------------------------------------
# 5. Detect Anomalies using Z-Score
# -------------------------------------------------------
df['Zscore'] = (df['Vehicle_Count'] - df['Vehicle_Count'].mean()) / df['Vehicle_Count'].std()

anomalies = df[df['Zscore'].abs() > 2]

print("\nDetected Anomalies (if any):")
print(anomalies)

# -------------------------------------------------------
# 6. Optional: Predict Peak Hours Using Regression
# -------------------------------------------------------
X = df[['Hour']]
y = df['Vehicle_Count']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print("\nRegression Model Accuracy:", accuracy)

# Predict future traffic for an hour
future_hour = 18   # example: 6 PM
prediction = model.predict([[future_hour]])
print(f"\nPredicted Traffic at Hour {future_hour}: {int(prediction[0])} vehicles")

# -------------------------------------------------------
# End of project
# -------------------------------------------------------
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load uploaded file
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print("\nDataset Loaded:")
print(df.head())

# -------------------------------------------------------
# FIX: Convert text violations into numeric counts
# -------------------------------------------------------
# Count number of violations for each hour
df['Violation_Count'] = df.groupby(df.columns[0])[df.columns[1]].transform('count')

# Keep only unique rows per hour for plotting
df_unique = df[[df.columns[0], 'Violation_Count']].drop_duplicates()

# -------------------------------------------------------
# 1. Plot violation count per hour
# -------------------------------------------------------
plt.figure(figsize=(10,5))
plt.plot(df_unique.iloc[:,0], df_unique['Violation_Count'], marker='o')
plt.title("Traffic Violations per Hour")
plt.xlabel(df.columns[0])
plt.ylabel("Number of Violations")
plt.grid(True)
plt.show()

# -------------------------------------------------------
# 2. Peak violation hour
# -------------------------------------------------------
peak = df_unique.loc[df_unique['Violation_Count'].idxmax()]
print("\nPeak Violation Hour:", peak[0])
print("Number of Violations:", peak[1])

# -------------------------------------------------------
# 3. Count types of violations
# -------------------------------------------------------
violation_stats = df[df.columns[1]].value_counts()
print("\nViolation Type Frequency:")
print(violation_stats)

# Plot violation types
plt.figure(figsize=(10,6))
violation_stats.plot(kind='bar')
plt.title("Violation Type Frequency")
plt.xlabel("Violation Type")
plt.ylabel("Count")
plt.show()

# -------------------------------------------------------
# 4. OPTIONAL: Encode violations numerically for ML models
# -------------------------------------------------------
label_encoder = LabelEncoder()
df['Violation_Code'] = label_encoder.fit_transform(df[df.columns[1]])

print("\nEncoded Violations Preview:")
print(df.head())
```
